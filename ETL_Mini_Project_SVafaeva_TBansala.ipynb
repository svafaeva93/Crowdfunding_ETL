### Extract the contacts.xlsx Data.

```python
# Read the data into a Pandas DataFrame. Use the `header=2` parameter when reading in the data.
contact_info_df = pd.read_excel('Resources/contacts.xlsx', header=2)
contact_info_df.head(3)
```

### Create the Contacts DataFrame 
---
**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

Then export the DataFrame as a `contacts.csv` CSV file.

### Option 1: Use Pandas to create the contacts DataFrame.

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
import json
# Create an empty list to hold the dictionaries
data_rows = []
for i, row in contact_info_df.iterrows():
    data_rows.append(row[0].split(","))
data_rows
# Print out the list of dictionaries
data_list = data_rows[1:]
print(data_list[0])
```

```python
result_list = []
for item in data_list:
    contact_dict = {}
    for field in item:
        if "contact_id" in field:
            contact_dict["contact_id"] = int(field.split(":")[1].strip())
        elif "name" in field:
            contact_dict["name"] = field.split(":")[1].strip()[1:-1]
        elif "email" in field:
            contact_dict["email"] = field.split(":")[1].strip()[1:-1]
    result_list.append(contact_dict)
print(result_list[0])
```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.
column_names = ['contact_id','name','email']
contact_info_new_df = pd.DataFrame(result_list, columns=column_names)
contact_info_new_df.head(3)
```

```python
# Check the datatypes.
contact_info_new_df.info()
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
# Split the name column into first_name and last_name columns
contact_info_new_df[['first_name', 'last_name']] = contact_info_new_df['name'].str.split(expand=True)
# Drop the contact_name column
contact_info_new_df = contact_info_new_df.drop('name', axis=1)
# removing the double quotes from email column
contact_info_new_df['email'] = contact_info_new_df['email'].str.replace('"', '')
# Print the updated dataframe
contact_info_new_df.head(3)
```

```python
# Reorder the columns
contact_info_new_df = contact_info_new_df[['contact_id', 'first_name', 'last_name', 'email']]
contact_info_new_df.head(3)
```

```python
# Check the datatypes one more time before exporting as CSV file.
contact_info_new_df.info()
```

```python
# Export the DataFrame as a CSV file. 
contacts_df_clean = contact_info_new_df.copy()
contacts_df_clean.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```

### Option 2: Use regex to create the contacts DataFrame.

```python
contact_info_df_copy = contact_info_df.copy()
# Removing the 0 row as it contained only contact_info word
contact_info_df_copy = contact_info_df_copy[~contact_info_df_copy.index.isin([0])]
contact_info_df_copy.head(3)
```

```python
# Extract the four-digit contact ID number.
contact_info_df_copy['contact_id'] = contact_info_df_copy.iloc[:, 0].str.extract(r'"contact_id":\s*(\d+)')
contact_info_df_copy.head(3)
```

```python
# Check the datatypes.
contact_info_df_copy.info()
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_id'].astype('int64')
# contact_info_df_copy.head(2)
contact_info_df_copy.info()
```

```python
# Extract the name of the contact and add it to a new column.
contact_info_df_copy['name'] = contact_info_df_copy.iloc[:, 0].str.extract(r'"name":\s*"([^"]+)"')
contact_info_df_copy.head(2)
```

```python
# Extract the email from the contacts and add the values to a new column.
contact_info_df_copy['email'] = contact_info_df_copy.iloc[:, 0].str.extract(r'"email":\s*"([^"]+)"')
contact_info_df_copy.head(2)
```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contact_info_df = contact_info_df_copy[['contact_id','name', 'email']].copy()
contact_info_df.head(3)
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
contact_info_df['first_name'] = contact_info_df['name'].str.extract(r'^(\w+)')
contact_info_df['last_name'] = contact_info_df['name'].str.extract(r'(\w+)$')
# Drop the contact_name column
contact_info_df = contact_info_df.filter(regex=r'^(?!name$).*')
contact_info_df.head(3)
```

```python
# Reorder the columns
contact_info_df = contact_info_df.filter(regex=r'^(contact_id|first_name|last_name|email)$')
contact_info_df.reset_index(drop=True, inplace=True)
contact_info_df.head(2)
```

```python
# Check the datatypes one more time before exporting as CSV file.
contact_info_df.info()
```

```python
# Export the DataFrame as a CSV file. 
contacts_df_clean = contact_info_df.copy()
contacts_df_clean.to_csv("Resources/regex_contacts.csv", encoding='utf8', index=False)
```
